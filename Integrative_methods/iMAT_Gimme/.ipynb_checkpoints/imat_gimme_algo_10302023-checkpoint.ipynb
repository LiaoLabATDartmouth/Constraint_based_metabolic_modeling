# Read in genome-scale metabolic model and set minimal medium with casamino acids
The casamino acids medium used in experiment contains all 20 amino acids except for tryptophan and essential micronutrients. For amino acids, we assume that their maximum uptake rates are all equal to 1 (mmol/hr/gDW). For micronutrients, we assume their uptake rates are unbounded (equal to 1000).

```python
# load Pseudomonas putida model deveoped by Nogales et al., 2017
model = cobra.io.load_json_model('SI3_iJN1411final_flux.json')
model.solver='gurobi' # you need to install gurobi solver first

# reset flux boudns to be between -1000 and 1000
for rxn in model.reactions:
    if rxn.lower_bound < -1000.0:
        rxn.lower_bound = -1000.0
    if rxn.upper_bound > 1000.0:
        rxn.upper_bound = 1000.0
```

```python
# set up the culture media
    
# first, close all uptake reactions
ex_rxns = [r for r in model.reactions if r.id.startswith('EX')]
for r in ex_rxns:
    r.lower_bound = 0
    
# second, reset lower bound of uptake fluxes only for nutrients in culture medium
CAA_min_media = {

    # trace elements
    'EX_pi_e_'     : -1000,     # phosphate
    'EX_fe3_e_'    : -1000,     # fe3+
    'EX_mn2_e_'    : -1000,     # mn2+
    'EX_zn2_e_'    : -1000,     # zn2+
    'EX_cu2_e_'    : -1000,     # cu2+
    'EX_ca2_e_'    : -1000,     # ca2+
    'EX_cl_e_'     : -1000,     # cl-
    'EX_cobalt2_e_': -1000,     # cobalt2+
    'EX_k_e_'      : -1000,     # k+
    'EX_mg2_e_'    : -1000,     # mg2+
    'EX_mobd_e_'   : -1000,     # mo6+
    'EX_na1_e_'    : -1000,     # na+
    'EX_ni2_e_'    : -1000,     # ni2+
    'EX_o2_e_'     : -1000,     # o2

    # amino acids
    'EX_leu_L_e_'  : -1,      
    'EX_ile_L_e_'  : -1,
    'EX_pro_L_e_'  : -1,
    'EX_trp_L_e_'  :  0,        # no tryptophan
    'EX_met_L_e_'  : -1,
    'EX_cys_L_e_'  : -1,
    'EX_val_L_e_'  : -1,
    'EX_asp_L_e_'  : -1,
    'EX_asn_L_e_'  : -1,
    'EX_ser_L_e_'  : -1,
    'EX_thr_L_e_'  : -1,
    'EX_ala_L_e_'  : -1,
    'EX_arg_L_e_'  : -1,
    'EX_gly_e_'    : -1,
    'EX_his_L_e_'  : -1,
    'EX_lys_L_e_'  : -1,
    'EX_phe_L_e_'  : -1,
    'EX_gln_L_e_'  : -1,
    'EX_glu_L_e_'  : -1,
    'EX_tyr_L_e_'  : -1
}
for rxn_id in CAA_min_media.keys():
    if rxn_id in model.reactions:
        rxn = model.reactions.get_by_id(rxn_id)
        rxn.lower_bound = CAA_min_media[rxn_id]
        rxn.upper_bound = 1000
    else:
        print("The reaction %s does not exist in model."%(rxn_id))
```

# Constrain biomass flux

```python
# read in estimated growth rates for each isolate/azt conc. combination
df_growth = pd.read_csv("exp_phase_growth_rate.csv")
df_growth.head()
```

```python
# calculate mean growth rate and its standard deviation for each combination
df_growth2 = pd.merge(
    df_growth.groupby(['Strain','Azt'])['GrowthRate'].mean(),
    df_growth.groupby(['Strain','Azt'])['GrowthRate'].std(),
    left_index=True,
    right_index=True,
    how='inner')
df_growth2.columns = ['GRmean','GRstd']
df_growth2 = df_growth2.reset_index()
df_growth2.index = ['%s_azt%d'%(s,a) for s,a in zip(df_growth2.Strain, df_growth2.Azt)]
df_growth2
```

```python
# create one model for each combination by constraining biomass flux between mean-std and mean+std
biom_constr_models = {}
biom_constr_samples = {}
for mid in df_norm_counts.columns:
    curr_model = deepcopy(model)
    mean_gr = df_growth2.loc[mid, 'GRmean']
    std_gr = df_growth2.loc[mid, 'GRstd']
    curr_model.reactions.BiomassKT2440_WT3.bounds = (mean_gr-std_gr, mean_gr+std_gr)
    biom_constr_models[mid] = curr_model
    
    print('sampling flux space of model %s...'%(mid))
    %time curr_samples = sample(curr_model, 10000, processes=4)
    biom_constr_samples[mid] = curr_samples
```

# Constrain model by transcriptomics
Information of transcriptomics data is converted to constraints on the ON/OFF reactions by maximizing the consistency between gene expression and flux. The solution can be found by the iMAT algorithm or the Gimme algorithm.

For reactions associated with multiple isozymes or one enzyme with multiple subunits, we determined their corresponding transcription levels by replacing “and” and “or” operators with “min” and “max” respectively in their gene-protein-reaction (GPR) rules. This is implemented in function or2min_and2max.

```python
def or2min_and2max(reaction, gene_values):
    """
    Parameters
    ----------
    reaction: reaction to be considered
    gene_values: a dict with gene ids as keys and their expression levels as values
    """
    
    string2parse = '('+reaction.gene_reaction_rule+')'
    
    # find the innermost paranthesis
    parens = [m for m in re.finditer(r"(\([^\(]*?\))",string2parse)]
    while len(parens)>0: # as long as a parenthesis can be found
        for paren in parens:
            extract = paren.group()
            content = extract[1:-1] # remove "(" and ")"
            if 'and' in content: # find the gene with mimumum expression value
                assert 'or' not in content
                genes = [g.strip() for g in content.split('and')]
                gene_exprs = [gene_values[g] for g in genes]
                substitute = genes[np.argmin(gene_exprs)]
            elif 'or' in content: # find the gene with maximum expression value
                assert 'and' not in content
                genes = [g.strip() for g in content.split('or')]
                gene_exprs = [gene_values[g] for g in genes]
                substitute = genes[np.argmax(gene_exprs)]
            else:
                substitute = content
            string2parse = string2parse.replace(extract, substitute)
            #print(string2parse)
        parens = [m for m in re.finditer(r"(\([^\(]*?\))",string2parse)]
    return string2parse, gene_values[string2parse]
    
# test
print("Reaction ID: ACLS")
print("Gene reaction rule: %s"%(model.reactions.get_by_id('ACLS').gpr))
print(or2min_and2max(model.reactions.get_by_id('ACLS'), {"PP_4679":10, "PP_4680":50, "PP_1157":30, "PP_3365":20}))
```

```python
def imat(model, expression_profile, condition, normalization, lowq=0.25, highq=0.75, epsilon=1.0, fraction_of_optimum=0.0):
    """
    Parameters
    ----------
    model: model for which fluxes will be inferred
    expression_profile: gene expression profiles
    condition: must be a column in the expression profile
    normalization: function to process GPR rules if a reaction is associated with multiple isozymes or one enzyme with multiple subunits
    lowq: the quantile to calcualte cutoff value for low expression genes
    highq: the quantile to calculate cutoff value for high expression values
    epsilon: a positive threshold to determine active flux associated with highly expressed reactions
    fraction_of_optimum: biomass constraint requires that the ratio of biomass flux to its maximum possible value must be equal or higher than "fraction_of_optimum"
    """
    
    # builds a dict that links gene IDs to their expression values
    assert condition in list(expression_profile.columns)
    expression_values = list(expression_profile[condition])
    gene_exp = dict(zip(expression_profile.index, expression_values))

    # binarize gene expression into "low", "middle", and "high"
    # note that we only use "low" and "high" genes to constrain fluxes
    # equivalent to np.percentile but q ranges from [0, 100]
    low_cutoff = np.quantile(expression_values, lowq, axis=0)
    high_cutoff = np.quantile(expression_values, highq, axis=0)
    
    # make a disctionary that maps reaction id to associated gene expression
    # note that we only constrain reactions using gene expression data when all associated genes can be found in expression profiles
    reaction_profile = {}
    for rxn in model.reactions:
        if len(rxn.genes) > 0 and all([identifier.id in expression_profile.index for identifier in rxn.genes]):
            _, reaction_profile[rxn.id] = normalization(rxn, {g.id: gene_exp.get(g.id) for g in rxn.genes})
        
    # flux variance analysis
    fva_res = flux_variability_analysis(model, list(reaction_profile.keys()), fraction_of_optimum=fraction_of_optimum)
    
    # add constraints that integrate on-off gene expression with fluxes
    y_variables = list()
    x_variables = list()
    constraints = list()
    for rid, expression in reaction_profile.items():
        if expression == 0.0:
            # reactions shut down completely
            model.reactions.get_by_id(rid).lower_bound = 0
            model.reactions.get_by_id(rid).upper_bound = 0
        elif expression >= high_cutoff:    
            # introduces two binary indicators y_pos and y_neg to control for the reaction activity in forward and reverse directions
            # note that when y_pos=1, flux>=epsilon; when y_pos=0, flux>=flux_min
            # when y_neg=1, flux<=-epsillon; when y_neg=0, flux<=flux_max
            # for a highly expressed gene, we want its catalyzed reaction to have either y_pos=1 or y_neg=1
            reaction = model.reactions.get_by_id(rid)
            y_pos = model.solver.interface.Variable("y_%s_pos" % rid, type="binary")
            y_neg = model.solver.interface.Variable("y_%s_neg" % rid, type="binary")
            y_variables.append([y_neg, y_pos])
            pos_constraint = model.solver.interface.Constraint(reaction.flux_expression + y_pos * (fva_res["minimum"][rid] - epsilon), lb=fva_res["minimum"][rid], name="pos_highly_%s" % rid)  
            neg_constraint = model.solver.interface.Constraint(reaction.flux_expression + y_neg * (fva_res["maximum"][rid] + epsilon), ub=fva_res["maximum"][rid], name="neg_highly_%s" % rid)
            constraints.extend([pos_constraint, neg_constraint])
        elif expression < low_cutoff:
            # introduces a single binary indicator x
            # note that when x=1, flux=0; when x=0, flux_min<=flux<=flux_max
            reaction = model.reactions.get_by_id(rid)
            x = model.solver.interface.Variable("x_%s" % rid, type="binary")
            x_variables.append(x)
            pos_constraint = model.solver.interface.Constraint((1 - x) * fva_res["maximum"][rid] - reaction.flux_expression, lb=0, name="x_%s_upper" % rid)
            neg_constraint = model.solver.interface.Constraint((1 - x) * fva_res["minimum"][rid] - reaction.flux_expression, ub=0, name="x_%s_lower" % rid)
            constraints.extend([pos_constraint, neg_constraint])

    len_xy = 0
    for variable in x_variables:
        model.solver.add(variable)
        len_xy += 1

    for variables in y_variables:
        model.solver.add(variables[0])
        model.solver.add(variables[1])
        len_xy += 1

    for constraint in constraints:
        model.solver.add(constraint)

    # solve for fluxes that maximize the consistency
    imat_objective = Add(*[(y[0] + y[1]) for y in y_variables]) + Add(*x_variables)
    model.objective = model.solver.interface.Objective(imat_objective, direction="max")
    solution = model.optimize()

    return model, reaction_profile, solution
```

```python
def gimme(model, expression_profile, condition, normalization, lowq=0.25, fraction_of_optimum=0.0):
    """
    Parameters
    ----------
    model: model for which fluxes will be inferred
    expression_profile: gene expression profiles
    condition: must be a column in the expression profile
    normalization: function to process GPR rules if a reaction is associated with multiple isozymes or one enzyme with multiple subunits
    lowq: the quantile to calcualte cutoff value for low expression genes
    fraction_of_optimum: biomass constraint requires that the ratio of biomass flux to its maximum possible value must be equal or higher than "fraction_of_optimum"
    """
    
    # constrain biomass flux
    fix_obj_constraint = model.solver.interface.Constraint(model.reactions.BiomassKT2440_WT3.flux_expression, lb=fraction_of_optimum * model.slim_optimize(), name="minimum_requirement_of_growth_rate")
    model.add_cons_vars(fix_obj_constraint)
        
    # builds a dict that links gene IDs to their expression values
    assert condition in list(expression_profile.columns)
    expression_values = list(expression_profile[condition])
    gene_exp = dict(zip(expression_profile.index, expression_values))

    # make a disctionary that maps reaction id to associated gene expression
    # note that we only constrain reactions using gene expression data when all associated genes can be found in expression profiles
    reaction_profile = {}
    for rxn in model.reactions:
        if len(rxn.genes) > 0 and all([identifier.id in expression_profile.index for identifier in rxn.genes]):
            _, reaction_profile[rxn.id] = normalization(rxn, {g.id: gene_exp.get(g.id) for g in rxn.genes})
    
    objective_terms = list()
    low_cutoff = np.quantile(expression_values, lowq, axis=0)
    coefficients = {rid:low_cutoff-exp if low_cutoff>exp else 0 for rid, exp in reaction_profile.items()}

    for rid, coefficient in coefficients.items():
        rxn = model.reactions.get_by_id(rid)
        if coefficient > 0:
            objective_terms.append(coefficient * (rxn.forward_variable + rxn.reverse_variable))
            
    gimme_objective = Add(*objective_terms)
    model.objective = model.solver.interface.Objective(gimme_objective, direction="min")
    solution = model.optimize()
    
    return model, reaction_profile, solution
```

```python
# run iMAT and Gimme
imat_models = {}
imat_solutions = {}
gimme_models = {}
gimme_solutions = {}
for mid in df_norm_counts.columns:
    
    ########################
    # get imat flux solution
    ########################
    imat_model = deepcopy(model)
    
    # set growth rate to experimental value
    exp_value = np.sum(imat_model.reactions.BiomassKT2440_WT3.bounds)/2
    imat_model.reactions.BiomassKT2440_WT3.bounds = (df_growth2.loc[mid,'GRmean'], df_growth2.loc[mid,'GRmean'])
    
    imat_model, _, imat_sol = imat(
        imat_model,
        expression_profile=df_norm_counts,
        condition=mid,
        normalization=or2min_and2max,
        lowq=0.25, 
        highq=0.75, 
        epsilon=0.1,
        fraction_of_optimum=0 # fraction of optimal growth should be set to 0 because biomass flux has been constrained by data
    )
    print("Run iMAT, condition %s, Objective value = %d"%(mid, imat_sol.objective_value))
    imat_models[mid] = imat_model
    imat_solutions[mid] = imat_sol
    
    #########################
    # get gimme flux solution
    #########################
    gimme_model = deepcopy(model)
    
    # set growth rate to experimental value
    gimme_model.reactions.BiomassKT2440_WT3.bounds = (df_growth2.loc[mid,'GRmean'], df_growth2.loc[mid,'GRmean'])
    
    gimme_model, _, gimme_sol = gimme(
        gimme_model,
        expression_profile=df_norm_counts,
        condition=mid,
        normalization=or2min_and2max,
        lowq=0.25, 
        fraction_of_optimum=0 # fraction of optimal growth should be set to 0 because biomass flux has been constrained by data
    )
    print("Run Gimme, condition %s, Objective value = %d"%(mid, gimme_sol.objective_value))
    gimme_models[mid] = gimme_model
    gimme_solutions[mid] = gimme_sol
```
